```python
import os
from time import sleep

print("Tugas SDA kelompok 5")
print('Nama anggota kelompok : ')
print('=======================================')
nama_kelompok = ['1. Dea Ananta              (2217051054)',
                 '2. Fernanda Pranata        (2217051146)',
                 '3. Muhammad Hildan Alfaris (2217051077)']
print(nama_kelompok[0])
print(nama_kelompok[1])
print(nama_kelompok[-1])
print('=======================================')

class keuangan:
    def __init__(self):
        self.pemasukan = []
        self.pengeluaran = []
        sleep(2)

    def tambah_pemasukan(self, jumlah):
        self.pemasukan.append(jumlah)
        sleep(2)
        
    def tambah_pengeluaran(self, jumlah):
        self.pengeluaran.append(jumlah)
        sleep(2)
        
    def hitung_total_pemasukan(self):
        return sum(self.pemasukan)
        sleep(2)
        
    def hitung_total_pengeluaran(self):
        return sum(self.pengeluaran)
        sleep(2)
        
    def hitung_saldo(self):
        return self.hitung_total_pemasukan() - self.hitung_total_pengeluaran()
        sleep(2)
        
    def daftar_keuangan(self):
        print("Daftar Pemasukan : ")
        for pemasukan in self.pemasukan:
            print(f"Rp. {pemasukan}")
        print(f"Total Pemasukan : Rp. {self.hitung_total_pemasukan()}")
        
        print()
        print("Daftar Pengeluaran : ")
        for pengeluaran in self.pengeluaran:
            print(f"Rp. {pengeluaran}")
            print(f"Total Pengeluaran : Rp. {self.hitung_total_pengeluaran()}")
        print()

        print(f"Jumlah Saldo : Rp. {self.hitung_saldo()}")
        print()
        sleep(2)
        
keuangan = keuangan()

print()
while True:
    print("===============================")
    print("      CATATAN KEUANGAN         ")
    print("===============================")
    print("Menu : ")
    print("1. Tambah Pemasukan")
    print("2. Tambah Pengeluaran")
    print("3. Daftar Keuangan")
    print("4. Keluar")
    pilihan = input("Masukkan Pilihan Anda : ")

    if pilihan == "1":
        jumlah = int(input("Masukkan jumlah Pemasukan : "))
        keuangan.tambah_pemasukan(jumlah)
        print("Pemasukan Anda Berhasil Ditambahkan!")
        print()
        
    elif pilihan == "2":
        jumlah = int(input("Masukkan jumlah Pengeluaran : "))
        keuangan.tambah_pengeluaran(jumlah)
        print("Pengeluaran Anda Berhasil Ditambahkan!")
        print()
        
    elif pilihan == "3":
        keuangan.daftar_keuangan()
        
    elif pilihan == "4":
        print("Terima kasih telah menggunakan program ini!")
        print()
        break
    
    else:
        print("Mohon maaf, pilihan tidak tersedia!!")
        print()
```
